```python
def load_lists_from_pickle_file(filepath):
    """
    Load a list of values from a pickle file at the specified filepath.

    Args:
        filepath (str): The filepath of the pickle file to load.

    Returns:
        list: The list of values loaded from the file.
    """
    with open(filepath, 'rb') as f:
        lists = pickle.load(f)
    return lists
```
